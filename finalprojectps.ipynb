```python
import subprocess
import numpy as np

#Welcome to the perfect panagram solver - one of the toughest tasks of the spelling bee

#Enter your 7 letters, and in a matter of seconds (minutes, several minutes)
# you will have the perfect panagram in the palm of your hand

def get_word_with_all_letters(letters, forbidden_words=None):
    if forbidden_words is None:
        forbidden_words = []
        
    smt_script = f'''
        (declare-sort Str 0)
        (set-option :produce-models true)
        (set-logic QF_LIA)

        (declare-const w String)

        (assert (distinct w ""))

        (assert (and { " ".join([ f'(str.contains w "{letter}")' for letter in letters ]) }))

    '''
    if forbidden_words:
        for fword in forbidden_words:
            smt_script += f'(assert (not (= w "{fword}")))'
            
    smt_script += '''
    (check-sat)
    (get-value (w))
    '''
    
    result = subprocess.run(['z3', '-in'], input=smt_script, capture_output=True, text=True)
    word_str = result.stdout.strip().split('\n')[-1]
    word = word_str[1:-1]  # remove quotes
    word = word.split()[1].split(')')[0].split('"')[1]
    word = list(word)

    #randomly spews out an upper case letter, this takes care of that
    for i in np.arange(len(word)):
        if word[i].isupper():
            word[i]=''
    
    for i in np.arange(len(word)):
        if word[i] not in letters:
            word[i]=''

    word = "".join(word)
    forbidden_words.append(word)
    
    return word
```

```python
web2lowerset = get_english_words_set(['web2'], lower=True)

letters = input("Enter 7 letters separated by spaces: ").split()
forbidden_words = []
panagram_found = False

keyword = get_word_with_all_letters(letters, forbidden_words)
print(keyword)

while panagram_found == False:
    if keyword in web2lowerset:
        print(keyword)
        panagram_found = True

    else:
        keyword = get_word_with_all_letters(letters, forbidden_words)
        print(keyword)
```
